# Outliers vs Noise Detection

* **Noise:** Incorrect Data Entry or have a Incorrect Measure
* **Outliers** is NOT Incorrect. have a High Distance with other Data and have low Frequency

in some case if you have outliers, that is awesome!

**Ex: the cholestoral Feature have a outlier data & that is suitable for you to easier reach the target!! so Why do you want to delete outliers?**

# Outliers Detection
